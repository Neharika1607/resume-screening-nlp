# Cleaning data ("Hashtags, special char,urls...")

```python
import re
def CleanResume(txt):
  CleanTxt= re.sub("http\S+", " ", txt)
  CleanTxt= re.sub("#\S+", " ", CleanTxt)
  CleanTxt= re.sub("@\S+", " ", CleanTxt)
  return CleanTxt
```
